```python
import os 
import pandas as pd
from dash import Dash, html
from flask import Response
import dag
```

```python
ABS_PATH = os.path.abspath("")
TRANSCRIPT_PATH = os.path.join(ABS_PATH, "transcriptions")
dfs = [pd.read_excel(os.path.join(TRANSCRIPT_PATH, fname)) for fname in os.listdir(TRANSCRIPT_PATH)]
df = dfs[-1]
```

```python
graph_html = dag.create_session_dag(df, name="lukas.hoelscher", session_id=9) 


app = Dash(__name__)
@app.server.route("/pyvis")
def pyvis():
    return Response(graph_html, mimetype="text/html")

app.layout = html.Div([
    html.Iframe(src="/pyvis", style={"width": "100%", "height": "800px", "border": "0"})
])

app.run(debug=True)
```
